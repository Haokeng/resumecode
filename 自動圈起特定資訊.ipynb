```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_finance as mpf
import seaborn as sns
import talib

# Download data from Yahoo Finance
df = yf.download('2330.TW', start='2022-10-01', end='2023-12-31')

# Convert index to string format with '%Y-%m-%d'
df.index = df.index.strftime('%Y-%m-%d')
#1030
sma_10 = talib.SMA(np.array(df['Close']), 10)
sma_30 = talib.SMA(np.array(df['Close']), 30)
sma_60 = talib.SMA(np.array(df['Close']), 60)
# Create figure and axes
fig = plt.figure(figsize=(24, 12))
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2, sharex=ax)

# Set x-axis ticks and labels
ax.set_xticks(range(0, len(df.index), 1))
ax.set_xticklabels(df.index[::1], rotation=45, ha='right')

# Plot candlestick chart
mpf.candlestick2_ochl(ax, df['Open'], df['Close'], df['High'], df['Low'],
                      width=0.6, colorup='r', colordown='g', alpha=0.75)

# Plot volume chart
mpf.volume_overlay(ax2, df['Open'], df['Close'], df['Volume'],
                   colorup='r', colordown='g', width=0.6, alpha=0.8)
#1030
ax.plot(sma_10, label='10')
ax.plot(sma_30, label='30')
ax.plot(sma_60, label='60')

plt.show()
```

```python
import pandas as pd
import yfinance as yf

# 2330.TWdata
data = yf.download("2330.TW", start="2016-01-01", end="2023-04-26")
from fbprophet import Prophet

# dataindex"ds""Adj Close""y"
data = data.reset_index()
data = data.rename(columns={"Date": "ds", "Adj Close": "y"})

# Prophet
model = Prophet()

# 
model.fit(data)

# dataframe365
future = model.make_future_dataframe(periods=365)

# 
forecast = model.predict(future)

# 
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
