```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import yfinance as yf
import mplfinance as mpf

# Download data from Yahoo Finance
df = yf.download('2330.TW', start='2021-01-01', end='2021-12-31')
df['Color'] = ['g' if df['Close'][i] >= df['Open'][i] else 'r' for i in range(len(df))]

# Find patterns
patterns = []
for i in range(2, len(df)-2):
    pattern = df.iloc[i-2:i+3]['Color'].values
    if pattern.tolist() == ['g', 'r', 'g', 'r', 'g']:
        if df.iloc[i-1]['Open'] > df.iloc[i-2]['Close'] and \
           df.iloc[i-1]['Open'] < df.iloc[i-1]['Close'] and \
           df.iloc[i-1]['Open'] < df.iloc[i]['Open'] and \
           df.iloc[i-1]['Open'] < df.iloc[i]['Close'] and \
           df.iloc[i-1]['Close'] > df.iloc[i]['Open'] and \
           df.iloc[i-1]['Close'] > df.iloc[i]['Close']:
            patterns.append((i-2, i+2))
    elif pattern.tolist() == ['r', 'g', 'r', 'g', 'r']:
        if df.iloc[i-1]['Open'] < df.iloc[i-2]['Close'] and \
           df.iloc[i-1]['Open'] > df.iloc[i-1]['Close'] and \
           df.iloc[i-1]['Open'] > df.iloc[i]['Open'] and \
           df.iloc[i-1]['Open'] > df.iloc[i]['Close'] and \
           df.iloc[i-1]['Close'] < df.iloc[i]['Open'] and \
           df.iloc[i-1]['Close'] < df.iloc[i]['Close']:
            patterns.append((i-2, i+2))

# Add Color column to dataframe
df['Color'] = ''
df.loc[df['Open'] < df['Close'], 'Color'] = 'g' # green
df.loc[df['Open'] > df['Close'], 'Color'] = 'r' # red

# Plot candlestick chart
mc = mpf.make_marketcolors(up='g', down='r', edge='inherit', wick='inherit')
s = mpf.make_mpf_style(marketcolors=mc)
mpf.plot(df, type='candle', style=s, volume=True, figratio=(12, 6), figscale=1.2)

# Draw rectangles around patterns
for pattern in patterns:
    rect = Rectangle((df.index[pattern[0]], df['Close'][pattern[0]]), 
                     (df.index[pattern[1]]-df.index[pattern[0]]), 
                     df['Close'][pattern[1]]-df['Close'][pattern[0]], 
                     linewidth=1, edgecolor='blue', facecolor='none')
    plt.gca().add_patch(rect)

# Show plot
plt.show()
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_finance as mpf
import seaborn as sns
import talib

# Download data from Yahoo Finance
df = yf.download('2330.TW', start='2023-01-01', end='2023-12-31')

# Convert index to string format with '%Y-%m-%d'
df.index = df.index.strftime('%Y-%m-%d')

# Compute candle colors
df['color'] = np.where(df['Open'] <= df['Close'], 'r', 'g')

# Compute differences between open and close prices
df['diff'] = abs(df['Open'] - df['Close'])

# Find green-red-green patterns where green diff is greater than red diff
grg_pattern = ((df['color'].shift(1) == 'g') & (df['color'] == 'r') & (df['color'].shift(-1) == 'g') &
               (df['diff'] > df['diff'].shift(1)) & (df['diff'] > df['diff'].shift(-1)))

# Create figure and axes
fig = plt.figure(figsize=(24, 12))
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2, sharex=ax)

# Set x-axis ticks and labels
ax.set_xticks(range(0, len(df.index), 1))
ax.set_xticklabels(df.index[::1], rotation=45, ha='right')

# Plot candlestick chart
mpf.candlestick2_ochl(ax, df['Open'], df['Close'], df['High'], df['Low'],
                      width=0.6, colorup='r', colordown='g', alpha=0.75)

# Plot volume chart
mpf.volume_overlay(ax2, df['Open'], df['Close'], df['Volume'],
                   colorup='r', colordown='g', width=0.6, alpha=0.8)

# Plot blue boxes around green-red-green patterns
for i in df[grg_pattern].index:
    mpf.plot(df.loc[i-1:i+1], type='candle', ax=ax, volume=ax2,
             addplot=[mpf.make_addplot([df['High'][i-1], df['High'][i], df['High'][i+1]],
                                       type='scatter', color='b', marker='s'),
                      mpf.make_addplot([df['Low'][i-1], df['Low'][i], df['Low'][i+1]],
                                       type='scatter', color='b', marker='s')])

plt.show()

```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpf
import seaborn as sns
import talib

# Download data from Yahoo Finance
df = yf.download('2330.TW', start='2023-01-01', end='2023-12-31')

# Convert index to datetime format
df.index = pd.to_datetime(df.index)

# Create figure and axes
fig = plt.figure(figsize=(24, 12))
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2, sharex=ax)

# Define function to identify green-red-green patterns
def find_grg_patterns(df):
    grg_pattern = []
    for i in range(1, len(df) - 1):
        if (df['Close'][i-1] > df['Open'][i-1] and df['Close'][i] < df['Open'][i] and df['Close'][i+1] > df['Open'][i+1]):
            if (df['Open'][i] - df['Close'][i] > df['Close'][i-1] - df['Open'][i-1] and df['Open'][i] - df['Close'][i] > df['Close'][i+1] - df['Open'][i+1]):
                grg_pattern.append(i)
    return grg_pattern

# Identify green-red-green patterns
grg_pattern = find_grg_patterns(df)

# Plot candlestick chart
mpf.plot(df, type='candle', ax=ax, volume=ax2, style='yahoo', 
         title='Taiwan Semiconductor Manufacturing Company Ltd. (2330.TW)',
         ylabel='Price (TW$)', ylabel_lower='Volume', 
         figscale=1.2, mav=(5, 10, 20), show_nontrading=False)

# Plot blue boxes around green-red-green patterns
for i in grg_pattern:
    mpf.plot(df.loc[df.index[i-1]:df.index[i+1]], type='candle', ax=ax, volume=ax2,
             addplot=[mpf.make_addplot([df['High'][i-1], df['High'][i], df['High'][i+1]],
                                       type='scatter', color='b', marker='s'),
                      mpf.make_addplot([df['Low'][i-1], df['Low'][i], df['Low'][i+1]],
                                       type='scatter', color='b', marker='s')])

plt.show()
```
